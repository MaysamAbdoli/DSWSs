```python
%matplotlib inline

import zipfile
import numpy as np
import pylab as plt
import pandas as pd
```

```python
zip_ref = zipfile.ZipFile('../datasets/south-park-dialogue.zip', 'r')
zip_ref.extractall('../datasets/')
```

```python
dialogue = pd.read_csv('../datasets/south-park-dialogue/All-seasons.csv')
```

```python
# dialogue[dialogue['Character']=='Stan']
```

```python
stan = dialogue[dialogue['Character']=='Stan']['Line'].values
```

```python
stan_words = []
for i in stan:
    for j in i.split(' '):
        if j!='\n':
            jp = j.replace('\n','')
            jp = jp.replace('?','')
            jp = jp.replace('!','')
            jp = jp.replace(',','')
            jp = jp.replace('.','')
            jp = jp.lower()
            stan_words.append(jp)
```

```python
# stan_words
```

```python
letters = 'bo'

possibilities = []
for i in stan_words:
    if i.startswith(letters):
        possibilities.append(i)
```

```python
possibilities
```

```python
poss,nums = np.unique(possibilities,return_counts=1)
sorted_indices = np.argsort(nums)[::-1]
```
